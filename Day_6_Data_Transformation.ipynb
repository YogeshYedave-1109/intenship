## Feature Scaling
- to make All feature having Same Scale without Changing thier internal Relation
- method - Normalization - Standerdization


- normalization is a Scaling technique in whoch values are shofted and resclaed so that they end up ranging between and 1 it is also known as Min-Max SCaling

```python
x=np.array([11,45,23,17,16,12,33])
x
```

```python
def normalize(a):
    norm=(a-a.min())/(a.max()-a.min())
    return norm
nx=normalize(x)
```

```python
nx
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Original Data")
plt.plot(x)
plt.subplot(1,2,2)
plt.title("Normalized Data")
plt.plot(nx)
plt.show()
```

```python
y=np.array([678,545,376,786,985,366,900])
ny=normalize(y)
ny
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Original Data")
plt.plot(y)
plt.subplot(1,2,2)
plt.title("Normalized Data")
plt.plot(ny)
plt.show()
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
x=df[['Age','EstimatedSalary']]
y=df['Purchased']
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.scatter(x['Age'],x['EstimatedSalary'],c=y)
plt.show()
```

```python
x.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
```

```python
type(x_scaled)
```

```python
x_new=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
x_new
```

```python
x_new.describe()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title("Original Data")
plt.scatter(x['Age'],x['EstimatedSalary'],c=y)
plt.subplot(1,2,2)
plt.title("Normalized Data")
plt.scatter(x_new['Age'],x_new['EstimatedSalary'],c=y)
plt.show()
```
