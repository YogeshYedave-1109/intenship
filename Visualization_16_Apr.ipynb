```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
x=np.array([1,25,66,78,99,54,93])
y=np.array([99,97,66,55,44,33,15])
z=np.array([73,62,58,44,39,32,45])
```

## line Plot

```python
plt.figure(figsize=(7,7))
plt.plot(x,y,marker='o',color='blue',mfc='r') # MFC= Marker Font color that is red 

plt.title("Line Plot")

plt.xlabel("First Value")
plt.ylabel("Second Value ")
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,y,marker='o',color='blue',mfc='r',ls='--',ms=8) # Ms= Marker Size   # ls = Line Style 
plt.title("Line Plot")

plt.xlabel("First Value")
plt.ylabel("Second Value ")
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,y,marker='o',color='blue',mfc='r',ls='--',ms=8) # Ms= Marker Size   # ls = Line Style 
plt.title("Line Plot")

plt.xlabel("First Value")
plt.ylabel("Second Value ")
plt.grid()
plt.show()
# marker => <,>,^,o,p,*,H
```

```python

plt.figure(figsize=(4,4))
plt.title("X & Z ")
plt.xlabel("X")
plt.ylabel("Z")
plt.plot(x,z ,marker='*',color='orange',mfc='b',ls='--',ms=6)
plt.plot(x,y,marker='o',color='blue',mfc='r',ls='--',ms=8) # Ms= Marker Size   # ls = Line Style 
plt.show()
```

## Sub plot

```python

plt.subplot(2,2,1)
plt.plot(x,y,marker='o',color='blue',mfc='r',ls='--',ms=8) # Ms= Marker Size   # ls = Line Style 
plt.title("Line Plot")

plt.xlabel("First Value")
plt.ylabel("Second Value ")
plt.subplot(2,2,2)
plt.plot(x,z ,marker='*',color='orange',mfc='b',ls='--',ms=6)
plt.title("Line Plot")

plt.xlabel("First Value")
plt.ylabel("Second Value ")
```

## Bar Graph
- Bar Graph
- color
- edgecolor
- width
- linewidth
- alpha

```python
students=[56,65,46,23]
courses=['Java','Python','C++','JavaSCript']
plt.figure(figsize=(7,7))
plt.bar(courses,students,color='orange',edgecolor='r',width=0.6,linewidth=3,alpha=0.5)
plt.title("Student Enrolled in Courses")
plt.xlabel("Courses")
plt.ylabel("Students")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y,marker='p')
plt.title("Scatter Plot")

plt.grid()
```

## Pie Chart
- it is used to show the contribution of every data point in dataset 

```python
plt.figure(figsize=(9,9))
plt.pie(students,labels=courses,colors=['red','green','yellow','orange'],
        autopct='%.2f%%',shadow=True,radius=0.9,labeldistance=1.2,startangle=30,explode=[0,0.1,0,0])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.bar(courses,students,color='orange',edgecolor='r',width=0.6,linewidth=3,alpha=0.5)
plt.title("Student Enrolled in Courses")
plt.xlabel("Courses")
plt.ylabel("Students")

plt.subplot(2,2,2)
plt.plot(x,z ,marker='*',color='orange',mfc='b',ls='--',ms=6)
plt.title("Line Plot")

plt.xlabel("First Value")
plt.ylabel("Second Value ")
plt.subplot(2,2,3)
plt.scatter(x,y,marker='p')
plt.title("Scatter Plot")

plt.grid()
plt.subplot(2,2,4)

plt.pie(students,labels=courses,colors=['red','green','yellow','orange'],
        autopct='%.2f%%',shadow=True,radius=0.9,labeldistance=1.2,startangle=30,explode=[0,0.1,0,0])

plt.savefig('fig')
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,hue='time');
plt.show()
```

```python
plt.xlabel("Tip Amount")
plt.ylabel("Frequency")


sns.histplot(df['tip'],kde=True)  # KDE is an line curve tha giev you the smooth curve ion plot 

plt.show()
```

```python
plt.title("Outliers in tip")
plt.xlabel("Day")
plt.ylabel("tip")
sns.boxenplot(x='day',y='tip',data=df)
plt.show()
```

```python
plt.title("Outliers in tip")
plt.xlabel("Day")
plt.ylabel("tip")
sns.boxenplot(x=df['day'],y=df['tip'])
plt.show()
```
